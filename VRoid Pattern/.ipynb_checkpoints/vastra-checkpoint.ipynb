```python
import cv2
from body_matrix import load, measure, draw
import torch
import numpy as np
from PIL import Image
import time

# Set up the device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load models
keypoints_model, keypoints_transform = load.keypoints_model(device)
segment_model, segment_transform = load.segment_model(device)

print("Models loaded successfully.")

# IP Webcam URL (replace with your IP Webcam address)
url = "http://192.168.29.199:8080/video"  # Update this URL
cap = cv2.VideoCapture(url)

if not cap.isOpened():
    print("Error: Could not open IP Webcam stream.")
else:
    print("IP Webcam stream opened successfully. Processing frames...")

    measurements_list = []
    no_person_frames = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture frame from IP Webcam.")
            break

        # Convert frame to RGB and PIL format
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame_pil = Image.fromarray(frame_rgb)

        # Extract measurements
        try:
            results = measure.find_real_measures(
                image_frame=frame_pil,
                device=device,
                keypoints_model=keypoints_model,
                keypoints_transform=keypoints_transform,
                segment_model=segment_model,
                segment_transform=segment_transform
            )

            # Check if results contain realistic measurements
            if results and len(results) >= 5:
                height, leg, hip, shoulder, markers, *extra_values = results

                # Validate the detected measurements
                if 100 < height < 250 and 50 < leg < 130 and 70 < hip < 150 and 40 < shoulder < 80:
                    measurements_list.append([height, leg, hip, shoulder])
                    print(f"Current measurements: Height: {height:.2f}, Leg: {leg:.2f}, Hip: {hip:.2f}, Shoulder: {shoulder:.2f}")

                    # When we have enough measurements, calculate the average
                    if len(measurements_list) >= 5:
                        avg_measurements = np.mean(measurements_list, axis=0)
                        height, leg, hip, shoulder = avg_measurements
                        visualized_frame = draw.visualize_measures(
                            height, leg, hip, shoulder, markers,
                            frame_pil, font_path
                        )

                        visualized_frame_cv = cv2.cvtColor(np.array(visualized_frame), cv2.COLOR_RGB2BGR)
                        cv2.imshow('Body Measurements', visualized_frame_cv)

                        print("Final Measurements:")
                        print(f"Height: {height:.2f}, Leg: {leg:.2f}, Hip: {hip:.2f}, Shoulder: {shoulder:.2f}")

                        # Save the final captured frame
                        captured_frame = visualized_frame_cv.copy()
                        break

                else:
                    no_person_frames += 1
                    print("No realistic measurements detected in this frame.")

            else:
                no_person_frames += 1
                print("No person detected in this frame.")

            # If no person is consistently detected, break the loop
            if no_person_frames > 10:
                print("No person consistently detected. Exiting measurement process.")
                break

        except Exception as e:
            print(f"Error during measurement extraction: {e}")

        # Display the live frame
        cv2.imshow('Webcam Feed', frame)

        # Exit if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release resources and close windows
    cap.release()
    cv2.destroyAllWindows()

    # Display the final captured measurements if obtained
    if 'captured_frame' in locals():
        cv2.imshow('Captured Measurements', captured_frame)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
```
